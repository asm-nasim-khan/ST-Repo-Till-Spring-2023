```python
p=input("Enter String:")
word =str(p)
length = len(word)
if(length<= 0):
    print("No String Found")
else:
    i = 0
    count = 0
    while i < length:
        if ord(word[i])>= 97:
            if ord(word[i])<= 122:
                count = count + 1
        i = i + 1
    if(count >= 4):
        print("Found")
    else:
        print("Not Found")
```
