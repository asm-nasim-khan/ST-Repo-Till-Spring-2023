```python
lst= []


while True:
  n=input("Enter:")
  if n!= "STOP":
    lst.append(int(n))
  else:
    break
done = []

for j in lst:
  count=0
  if j not in done:
    for k in lst:
      if j == k:
        count+=1
    print(j,"-",count,"times")
    done.append(j)
```
