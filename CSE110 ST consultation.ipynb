```python
def capitalization(para):
    result= ""
    for i in range (len(para)):
      if i <= len(para)-2:
        if i == 0:
          y = para[i].upper()
          result += y
        elif para[i] == "." or para[i] == "?" or para[i] == "!": 
          result += para[i] + ' '
          y = para[i+2].upper()
          result +=y
        elif para[i] == " " and para[i-1]== "." and para[i+1]!=" ":
          pass
        elif (para[i-2] == "." or para[i-2] == "?" or para[i-2] == "!") and i>2:
          pass
        elif para[i]== "i":
          if para[i- 1] == para[i +1 ]:
            y = para[i].upper()
            result += y
          else:
            result += "i"
        elif para[-1] == '.':
          result += para[i]
        else:
          result += para[i]
      else:
        result += para[i]
    return result

    
str1 = input('Enter: ')
print(capitalization(str1))
```
